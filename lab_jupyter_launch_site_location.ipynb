Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordenadas de los 4 sitios de lanzamiento con sus costas más cercanas
launches_data = [
    {
        'name': 'CCAFS LC-40',
        'launch_lat': 28.562302,
        'launch_lon': -80.577356,
        'coast_lat': 28.56143,  # Punto intermedo
        'coast_lon': -80.56772,
        'color': 'blue'
    },
    {
        'name': 'CCAFS SLC-40',
        'launch_lat': 28.563197,
        'launch_lon': -80.576820,
        'coast_lat': 28.56328,  # Punto más al este
        'coast_lon': -80.56802,
        'color': 'red'
    },
    {
        'name': 'KSC LC-39A',
        'launch_lat': 28.573255,
        'launch_lon': -80.646895,
        'coast_lat': 28.58041,  # Punto más al interior
        'coast_lon': -80.64252,
        'color': 'green'
    },
    {
        'name': 'VAFB SLC-4E',
        'launch_lat': 34.632834,
        'launch_lon': -120.610745,
        'coast_lat': 34.63588,  # Costa oeste
        'coast_lon': -120.62505,
        'color': 'orange'
    }
]

# Crear las 4 líneas automáticamente
for launch in launches_data:
    # Calcular distancia
    distance = calculate_distance(
        launch['launch_lat'], 
        launch['launch_lon'],
        launch['coast_lat'], 
        launch['coast_lon']
    )
    
    # Coordenadas del lanzamiento y costa
    launch_coord = [launch['launch_lat'], launch['launch_lon']]
    coast_coord = [launch['coast_lat'], launch['coast_lon']]
    
    # Crear marker en el punto de costa
    folium.Marker(
        coast_coord,
        icon=folium.DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>✗</b></div> {distance:.2f} KM'
        )
    ).add_to(site_map)
    
    # Crear línea entre lanzamiento y costa
    lines = folium.PolyLine(
        locations=[launch_coord, coast_coord],
        weight=2,
        color=launch['color'],
        opacity=0.7,
        popup=f"{launch['name']}: {distance:.2f} km"
    ).add_to(site_map)
    
    print(f"{launch['name']}: {distance:.2f} km hasta la costa")
    
site_map.add_child(lines)
```

```python
launches_data2 = [{
        'name': 'VAFB SLC-4E',
        'launch_lat': 34.632834,
        'launch_lon': -120.610745,
        'lineas': [
            {'tipo': 'costa', 'lat': 34.63588, 'lon': -120.62505, 'color': 'blue'},
            {'tipo': 'tren', 'lat': 34.6358, 'lon': -120.62347, 'color': 'grey'},
            {'tipo': 'carretera', 'lat': 34.63458, 'lon': -120.61863, 'color': 'orange'},
            {'tipo': 'ciudad', 'lat': 34.64479, 'lon': -120.45959, 'color': 'black'}],
        
    }
]
for launch in launches_data2:
    launch_coord = [launch['launch_lat'], launch['launch_lon']]
    # Calcular distancia
    for linea in launch['lineas']:
        distance = calculate_distance(
            launch['launch_lat'], 
            launch['launch_lon'],
            linea['lat'], 
            linea['lon']
        )
        linea_coord = [linea['lat'], linea['lon']]
        
        folium.Marker(
            linea_coord,
            icon=folium.DivIcon(
                icon_size=(20,20),
                icon_anchor=(0,0),
                html=f'<div style="font-size: 12; color:#d35400;"><b>✗</b></div> {distance:.2f} KM'
            )
        ).add_to(site_map)
        
        # Crear línea entre lanzamiento y costa
        lines = folium.PolyLine(
            locations=[launch_coord, linea_coord],
            weight=4,
            color=linea['color'],
            opacity=0.7,
            popup=f"{launch['name']} {linea['tipo']}: {distance:.2f} km"
        ).add_to(site_map)
        
        print(f"{launch['name']} {linea['tipo']}: {distance:.2f} km hasta la costa")
    
site_map.add_child(lines)
```
